```python
#Generation of clock pulse from sinusoid

from numpy import*
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
t=linspace(0,1,1000);
x=sin(2*pi*5*t)

def sqp(x):
    p=zeros_like(x);
    for n in range(x.size):
        p[n]=2*(x[n]>0)-1;
    p[0]=0;    
    return p

r=fft(sqp(x));
s=fft(x);
print(abs(r[-1]),abs(s[-1]));
n=sqp(x).size;

f1=plt.figure();
plt.plot(t,sqp(x));
plt.plot(t,x);
f2=plt.figure();
plt.plot(fftfreq(n,0.1),abs(r));
plt.plot(fftfreq(n,0.1),abs(s));
plt.show();
```

```python
import matplotlib.pyplot as plt
class sin_gen:
    
    def __init__(self,clk,rstb):
        self.clk=clk
        self.rstb=rstb
        self.seq=np.zeros(len(self.clk))
        self.out=np.zeros(len(self.clk))
        
    def sine_gen(self): #Acts as a ROM and stores the whole sequence for Sine
        a=[2,3,4,9,10,13,17,21,23,32,40]
        seq1=np.ones(64)
        for i in range(0,len(a)):
            seq1[a[i]]=0
        seq2=flip(seq1)
        seq3=np.ones(len(seq1))-seq1
        seq4=flip(seq3)
        seq=np.concatenate((seq1,seq2,seq3,seq4)) 
        j=0
        for i in range(0,len(self.clk)): #makes the sequence length(256) repeat according to the clk length
            if j+1<len(seq):
                self.seq[i]=seq[j]
            else:
                j=-1
            j=j+1    
        return self.seq
    
    
    def cos_gen(self): #Acts as a ROM and stores the whole sequence for Cosine
        a=[2,3,4,9,10,13,17,21,23,32,40]
        seq4=np.ones(64)
        for i in range(0,len(a)):
            seq4[a[i]]=0
        seq1=flip(seq4)
        seq2=np.ones(len(seq4))-seq4
        seq3=flip(seq2)
        seq=np.concatenate((seq1,seq2,seq3,seq4)) 
        j=0
        for i in range(0,len(self.clk)): #makes the sequence length(256) repeat according to the clk length
            if j<len(seq):
                self.seq[i]=seq[j]
            else:
                j=-1
                
            j=j+1    
        return self.seq
   
    
    def drive_sine(self): #drive circuit with asynchronous reset bit, i.e if rstb=0--> reset, else: set
         
        j=0
        for i in range(0,len(self.clk)-1):
            if self.rstb[i]==1:
                if not self.clk[i] and self.clk[i+1]==1: 
                    self.out[i]=self.sine_gen()[j]
                else: 
                    self.out[i]=self.out[i-1]
                    j=j-1
                j=j+1 
            else:
                j=0 #makes the sequence start from scratch after reset
                self.out[i]=0
        
        return self.out
    
    def drive_cos(self): #drive circuit with asynchronous reset bit, i.e if rstb=0--> reset, else: set
         
        j=0
        for i in range(0,len(self.clk)-1):
            if self.rstb[i]==1:
                if not self.clk[i] and self.clk[i+1]==1: 
                    self.out[i]=self.cos_gen()[j]
                else: 
                    self.out[i]=self.out[i-1]
                    j=j-1
                j=j+1 
            else:
                j=0 #makes the sequence start from scratch after reset
                self.out[i]=0
        return self.out
             
#testing            
clk=np.zeros(256*2) 
clk[0::2]+=1
rstb=np.ones(256*2)
rstb[0:20]=0
rstb[400:600]=0
a=sin_gen(clk,rstb)
#plt.plot(clk)
f1=plt.figure()
plt.plot(a.sine_gen())
plt.plot(rstb/2)
#print(a.sequence_gen())
#print(a.drive())
#print(clk)
f2=plt.figure()
plt.plot(a.drive_sine())
plt.plot(rstb/2)

f3=plt.figure()
plt.plot(a.cos_gen())
plt.plot(rstb/2)
print(a.cos_gen())

f4=plt.figure()
plt.plot(a.drive_cos())
plt.plot(rstb/2)
        
```

```python
#Optimization Task

from numpy import *
import matplotlib.pyplot as plt
from scipy.fft import fft    #import scipy.fft after numpy import to prevent ('module' not callable error)

class Optimization:
    
    def __init__(self,n,en): #n is the total number of bits for the signal sequence 
        self.n=n             #en is the no. of freqencies to be included in for calulating the ratio; en+1 because first 
        self.en=en+1           #bin of an fft is not frequency but average energy of whole fft. So FF is the second bin and 1st
        self.quart=int(n/4)    #harmonic at 2nd bin
                            #self.quart is the quarter of the signal sequence
                            #int() converts the float into integer
    
    def ref_signal(self):    #generates the reference signal:256 bit optimized signal we already have
        zero_bits=[2,3,4,9,10,13,17,21,23,32,40]
        sequence=zeros(64*4)
        sequence[0:(64)]=ones(64)
        
        for i in range(0,len(zero_bits)):
            sequence[zero_bits[i]]=0
            sequence[64:2*64]=flip(sequence[0:(64)])
            sequence[2*64:4*64]=ones(2*64)-sequence[0:2*64]
       
        return sequence
    
    def opt_signal(self):  #generates all possible signal sequences (2^n); n is the no. of bits in a quarter
        n=self.quart      #self.quart=no. of bits in a quarter
        arr=zeros((pow(2,n),n)) #change data type to reduce memory. By default, array is stored as float64
        for i in range(1,pow(2,n)): 
            arr[i,0]=1-arr[i-1,0]
            for j in range(1,n):
                if not arr[i,j-1] and arr[i-1,j-1]:
                    arr[i,j]=1-arr[i-1,j]
                else:
                    arr[i,j]=arr[i-1,j]
        
        arr_seq=zeros((pow(2,n),4*n))           
        for i in range(0,pow(2,n)):# to generate full sequence array from array of quarters
            arr_seq[i,0:n]=arr[i,:]
            arr_seq[i,n:2*n]=flip(arr[i,:])
            arr_seq[i,2*n:4*n]=1-arr_seq[i,0:2*n]
        #savetxt('signal_array.csv',arr_seq)
        return arr_seq  #returns a [2^n,4n] array; 
    
    def fft_array(self):                #gives the fft of all the possible signal sequences and the reference signal
        signal_arr=self.opt_signal()
        n=self.quart
        fft_ref=fft(self.ref_signal())
        fft_orig=abs(fft_ref)/amax(abs(fft_ref)) #0th bin of fft
        fft_arr=zeros((pow(2,n),4*n))
        for i in range(0,pow(2,n)):
            fft_arr[i,:]=abs(fft(signal_arr[i,:]))/amax(abs(fft(signal_arr[i,:])))
        return fft_orig,fft_arr   #fft_orig is the fft of reference signal
                                  #fft_arr is the fft array of all possible signal sequences  
   
    def energy_array(self):     #Returns the energy ratio: Energy(Fundamental Frequency: FF)/Energy(FF+closeby harmonics)
        n=self.quart
        en=self.en
        fft_ref=self.fft_array()[0]
        fft_arr=self.fft_array()[1]
        TE_ref=0
        TE_opt=zeros((pow(2,n),1))
        ER_opt=zeros_like(TE_opt)
        #Maximizing energy for first 10(en) bin outputs of fft as our filter has a order 3, so from cut off freq of 10kHz 
        #to 100kHz, it will itself drop down by -60dB (as freq goes to 10 times of cut off freq)
        #Neglect the first bin as it is not frequency but avaerage energy across all frequencies in the bin
        for i in range(1,en):#TE: Total Energy; ER: Energy ratio
            TE_ref+=pow(fft_ref[i],2) #The 0th bin of fft represents the average energy of the fft, so start from 1 instead of 0
        ER_ref=pow(fft_ref[1],2)/TE_ref #The first half of the fft repeats in the second half
        
        for i in range(0,pow(2,n)):
            for j in range(1,en):
                TE_opt[i]+=pow(fft_arr[i,j],2)
                
            ER_opt[i]=pow(fft_arr[i,1],2)/TE_opt[i]
        
        return ER_ref, ER_opt    #ER_ref is the energy ratio of reference signal
                                 #ER_opt is the array with energy ratio across all the possible sequence cases
 
    def opt_out(self):           #returns optimal sequence, optimal fft, optimal energy ratio and the index
        signal_seq_arr=self.opt_signal()         #index: it represents the index of optimal sequence out of all possible cases
        signal_fft_arr=self.fft_array()[1]
        ER_arr=self.energy_array()[1]
        ER_max=amax(ER_arr)
        index=zeros(len(ER_arr))
        for i in range(0,len(ER_arr)):
            if ER_max==ER_arr[i,0]:
                index[i]=i
        index_max=amax(index)
        optimal_sequence=signal_seq_arr[int(index_max),:]
        optimal_fft=signal_fft_arr[int(index_max),:]
        optimal_ER=ER_arr[int(index_max)]
        return optimal_sequence,optimal_fft,optimal_ER,index_max
        
class regression_bits:   #does regression in no. of closeby harmonics and the bit-sequence size
    
    def __init__(self,start,stop,step,har_start,har_stop,har_step):
        self.start=start
        self.step=step
        self.stop=stop
        self.har_start=har_start
        self.har_stop=har_stop
        self.har_step=har_step
        
    def regression(self):
        start=self.start
        stop=self.stop
        step=self.step
        har_start=self.har_start
        har_stop=self.har_stop
        har_step=self.har_step
        
        N=int((stop-start)/step+1) #N is the number of cases of Sequence bit variation
        H=int((har_stop-har_start)/har_step+1) #H is the no of cases for harmonic variation
        ER_ref=zeros(H)
        ER_arr=zeros((H,N))
        index=zeros(N)
    
        for k in range(H):                #array rows: variation in harmonic number
            for i in range(N):           #array col: variation in number of bits in sequence
                Opt=Optimization(start+(i*step),har_start+(k*har_step))
                ER_arr[k,i]=Opt.opt_out()[2]
                index[i]=start+(i*step)
                ER_ref[k]=Opt.energy_array()[0]
            #f1=plt.figure()
            plt.stem(index,ER_arr[k,:],linefmt='r',markerfmt='o',label=['Harmonics: %s' %(k*har_step+har_start)])
            plt.xlabel('No of bits in the Sequence')
            plt.ylabel('Energy_Ratio (Energy_FF/T.Energy_harmonics)')
            plt.title('Variation in Energy ratios: Stem plot')
            plt.legend()
            plt.grid()
        plt.show()
        
        for k in range(H):
            #f2=plt.figure()
            plt.plot(index,ER_arr[k,:],'--',label=['Harmonics: %s' %(k*har_step+har_start)])
            plt.axis([start-5,stop+5,0.5,1.2])
            plt.xlabel('No of bits in the Sequence')
            plt.ylabel('Energy_Ratio (Energy_FF/T.Energy_harmonics)')
            plt.title('Variation in energy ratio with No of Bits in Sequence and harmonics')
            plt.grid()
            plt.legend()
        plt.show()
        
        savetxt('Energy_arr.csv',ER_arr)
        #This snippet gives out the data of the best sequence in the whole regression
        
        har_input=int(input("Either enter '0' or Do the analysis of the Energy_arr file and enter the value of optimal Harmonics you choose:\n"))
        bits_input=int(input("Either enter '0' or Do the analysis of the Energy_arr file and enter the value of optimal Bits in sequence required:\n"))
        
        if har_input==0 and bits_input==0:
            ER_max_index=unravel_index(argmax(ER_arr,axis=None),ER_arr.shape) 
            harmonic=har_start+(ER_max_index[0]*har_step)
            Bits=start+(ER_max_index[1]*step)
        else:
            Bits=bits_input
            harmonic=har_input
            
        Opt_seq=Optimization(Bits,harmonic) #Running the process again to gather details of the best sequence
        fft_ref_max=Opt_seq.fft_array()[0] #refrence fft for best ER ratio of signals and harmonics
        fft_best_seq=Opt_seq.opt_out()[1]
        print(ER_arr) #gives ER array with rows: harmonics variation; col: No of bits in sequence variation
        print(ER_ref) #gives ER of reference signal for comparison
        
        #print fft comparison
        plt.plot(fft_ref_max,label='fft_reference')
        plt.axis([-0.1,1.1,-1,50])
        plt.plot(fft_best_seq,'y--',label='fft_best_sequence')
        plt.axis([-0.1,1.1,-1,50])
        plt.xlabel('No of bins')
        plt.ylabel('Normalised Absolute FFT')
        plt.title('Comparison in the FFT between Reference and Best Sequence')
        plt.grid()
        plt.legend()
        
        #prints relevant data in csv format for further analysis              
        savetxt('OPT_seq.csv',Opt_seq.opt_out()[0])
        savetxt('OPT_fft.csv',Opt_seq.opt_out()[1])
        savetxt('OPT_ER.csv',Opt_seq.opt_out()[2])
        #savetxt('OPT_index.csv',Opt_seq.opt_out()[3])              

        
        
        
        
        
'''       
x=Optimization(72,11)
ref_signal=x.ref_signal()
t=x.fft_array()[0].dtype
print(t)
f1=plt.figure()
plt.plot(ref_signal)

f2=plt.figure()
opt_seq=x.opt_out()[0]
plt.plot(opt_seq)
print(opt_seq)

f3=plt.figure()
opt_fft=x.opt_out()[1]
plt.stem(opt_fft)

f4=plt.figure()
ref_fft=x.fft_array()[0]
plt.plot(ref_fft)
plt.plot(opt_fft,'y--')
print(opt_fft)

opt_ER=x.opt_out()[2]
print(opt_ER)
a=x.energy_array()
print((a[0]))
print(amax(a[1]))

opt_index=x.opt_out()[3]
print(opt_index)
print(amax(opt_index))
'''

#Class Definition: regression_bits: Regression of No of Bits in sequence and no of harmonics
#regression_bits(Start,Stop,Step,Harmonic_start,Harmonic_stop,Harmonic_step)
#Start: Start bit size of sequence
#Stop: Max bit size of sequence
#Step: step increment(keep a multiple of 4)

a=regression_bits(64,80,4,2,8,2) #Harmonic_step=2 taken as alternate entries are zero in fft. So no change is energy for
a.regression()                   #adjacent no. of harmonics


```

```python
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
from threading import Thread
from multiprocessing import Process


"""
Steps for optimization:
1. For n bit sequence, we take n/4 bits and then create the full sin wave by muxing it
2. To generate the sequence: create bin(2**(n/4)) and then subtract 1 from it in each iteration until it goes to 0. This new sequence will be what we need for operation
"""

class optimize_sequence:

    def __init__(self, n, thd_min, init_n=0, end_n=1):
        self.bits = int(n/4)
        self.init_n = init_n
        self.end_n = end_n
        self.num_seq=18432

    def exec_thread(self):
        for i in range(self.init_n, self.end_n):
            self.num_seq = i
            seq = self.complete_sequence()
            self.get_thd(seq, thd_min)

    def generate_nth_quarter(self):
        seq = np.zeros(self.bits, dtype = np.int16)
        binary_seq = bin((2**self.bits)-1-self.num_seq)[2:]
        for i in range(len(binary_seq)):
            seq[len(seq)-len(binary_seq)+i] = binary_seq[i]
        return seq

    def complete_sequence(self):
        seq = self.generate_nth_quarter()
        seq_first = np.concatenate((seq,seq[::-1]), axis=None)
        seq_inv = []
        for i in seq_first:
            if i ==1:
                seq_inv.append(0)
            else:
                seq_inv.append(1)
        return np.concatenate((seq_first,seq_inv), axis=None)

    def find_fft(self, seq):
        return np.fft.fft(seq)

    def get_thd(self, seq, thd_min):
        fft = self.find_fft(seq)
        # print(fft)
        thd_num = np.sum(np.square(np.abs(fft[2:])))
        thd_den = np.abs(fft[1])
        thd = thd_den/np.sqrt(thd_num) ## Note that this is the inverse of thd
        thd_min[self.num_seq] = thd

if __name__ == "__main__":

    bit_length = 64
    min_ = 0
    max_ = int(2**(bit_length/4)-1)
    thd_min = np.zeros(int(2**(bit_length/4)),dtype=np.float16)
    for i in range(min_, max_, 100000):
        print(i)
        threads = []
        for k in range(i,i+100000,1000):
            if k+1000<max_:
                o = optimize_sequence(bit_length,thd_min,k,k+1000)
                threads.append(Thread(target=o.exec_thread, args=()))
            else:
                o = optimize_sequence(bit_length,thd_min,k,max_)
                threads.append(Thread(target=o.exec_thread, args=()))
        for th in range(0,len(threads),8):
            if th+8<len(threads):
                end = th+8
            else:
                end = len(threads)
            for j in range(th, end):
                threads[j].start()
            for j in range(th,end):
                threads[j].join()
    print("Waiting here")
    print(np.argmax(thd_min))
    obj = optimize_sequence(bit_length, thd_min, min_, max_)
    # print(obj.complete_sequence())
#     arr = obj.complete_sequence()
#     plt.stem(np.abs(np.fft.fft(arr)/np.max(np.fft.fft(arr)))[1:])
```
